## Stacks

<ul>
<li>LIFO Approach (Last in First Out)
<li> Stack is a Linear Data Structure
</ul>


### Operations in Stack
1. Push - To Insert an Element in Stack, inserting at front of a list
2. Pop  - To Remove an Element in Stack, removing from front of a list
3. Top  - Returns Top Element of the Stack
4. IsFull - Checks if Stack is Full
5. IsEmpty - Checks if Stack is Empty

```python
class Stack:
    def __init__(self,max_size):
        self.__max_size = max_size
        self.__elements = [None]*self.__max_size
        self.__top = -1
    
    def is_full(self):
        if self.__top == self.__max_size-1:
            return True
        return False
    
    def is_empty(self):
        if self.__top==-1:
            return True
        return False
    
    def push(self,element):
        if self.is_full()==True:
            print("Stack is Full")
        else:
            self.__top+=1
            self.__elements[self.__top]=element
    
    def pop(self):
        if self.is_empty():
            print("Stack is Empty")
        else:
            data = self.__elements[self.__top]
            self.__top-=1
            return data
            
    def peek(self):
        if self.is_empty():
            print("Stack is Empty")
        else:
            print(self.__elements[self.__top::-1])
            
    def maxsize(self):
        print(self.__max_size-1)
        
```

```python
stack.push(1)
stack.peek()
```

## Queues



<ul>
    <li> FIFO Approach (First In First Out)
    <li> Linear Data Structure

### Operations in Queues
1. isempty() - Determines if the Queue is Empty
2. isfull() - Determines if the queue is full in case of bounded size queue
3. front - Returns Element at the Front of The Queue
4. Enqueue - Inserts an Element at the Rear
5. Dequeue - Removes Element from the Front

```python
class Queue:
    def __init__(self,max_size):
        self.__max_size = max_size
        self.__elements = [None]*self.__max_size
        self.__rear = -1
        self.__front = 0
        
    def isfull(self):
        if self.__rear == self.__max_size-1:
            return True
        return False
    
    def isempty(self):
        if self.__front>self.__rear:
            return True
        return False
    
    def enqueue(self,element):
        if self.isfull():
            print("Queue Full")
        else:
            self.__rear+=1
            self.__elements[self.__rear]=element
            
    def dequeue(self):
        if self.isempty():
            print("Queue is Empty")
        else:
            data = self.__elements[self.__front]
            self.__front+=1
            return data
            
    def display(self):
        if self.isempty():
            print("Queue Empty")
        else:
            print(self.__elements[self.__front:self.__rear+1])
            
    def maxsize(self):
        return self.__max_size
```

```python
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
```

### Given a Queue of whole numbers. Write a Python Function to return a new queue whihc contains evenly divisible numbers

#### A Number is said ot be evenly divisible if it is divisble by all the numbers in given rance without any remainder. Consider the range to be from 1 to 10.
Assume that there will always be few elements in the input queue which are evenly divisible by numbers in mentioned range

```python
n = int(input())
ipq = Queue(n)
for i in range(n):
    ipq.enqueue(int(input()))
opq = Queue(10)
for i in range(n):
    ele = ipq.dequeue()
    cnt=0
    for j in range(1,11):
        if ele%j==0:
            cnt+=1
    if cnt==10:
        opq.enqueue(ele)
        
opq.display()
```

```python
class Node:
    def __init__(self,dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None
    
    def listprint(self,sep=''):
        printval = self.headval
        while printval is not None:
            print(printval.dataval,end=sep)
            printval = printval.nextval
            
    def insertend(self,node):
        if self.headval == None:
            self.headval = node
            return
        laste = self.headval
        while(laste.nextval):
            laste = laste.nextval
        laste.nextval=node
    
    def insertstart(self,node):
        node.nextval = self.headval
        self.headval = node
```

```python
linked_list = SLinkedList()
linked_list.headval = Node('Mon')
e2 = Node("Tue")
e3 = Node ("Wed")

#Linking First Node to Second Node
linked_list.headval.nextval = e2

#Linking 2nd Node to Third Node
e2.nextval = e3
```

Rules
1. Replace '*' or '/' by a single space
2. In case of two consecutive occurences of '*' or '/' replace those two occurence by a single spaace and convert the next character to uppercase

Assume that
There will not be more than two consecutive occurences of '*' or '/' The linked list will always end with an alphabet

Sample Input
<p> A,n,*,/,a,p,p,l,e,*,a,/,day,*,*,k,e,e,p,s,/,*,a,/,/,d,o,c,t,o,r,*,A,w,a,y </p>

Output

```python
l = list(input().split(','))
```

```python
single = SLinkedList()
i=0
while i<len(l):
    try:
        if (l[i]=='*' and l[i+1]=='/') or (l[i]=='/' and l[i+1]=='*') or (l[i]=='*' and l[i+1]=='*') or (l[i]=='/' and l[i+1]=='/'):
            single.insertend(Node(' '))
            i+=2
            l[i] = l[i].upper()
            continue
        elif l[i]=='*' or l[i]=='/':
            single.insertend(Node(' '))
        else:
            single.insertend(Node(l[i]))
        i+=1
    except:
        break
```
